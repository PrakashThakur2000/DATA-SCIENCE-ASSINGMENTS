# MODEL BULDING

```python
import statsmodels.formula.api as smf
```

```python
model=smf.ols('Salary~YEXP',data=df).fit()
```

```python
model.params
```

```python
model.pvalues
```

```python
model.rsquared
```

```python
model.rsquared_adj
```

# Testing of model

```python
test=pd.DataFrame({
    'YEXP':[1.1,2.0,3.2,4.1,5.6]
})
test
```
