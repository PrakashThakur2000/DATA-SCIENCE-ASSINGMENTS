# K-Means clustering

```python
data_set=pd.read_csv("crime_data.csv")
data_set
```

```python
data_set.drop(columns="Unnamed: 0",inplace=True)
data_set
```

```python
# data_set visualization
```

```python
sns.scatterplot(data_set)
```

```python
# feature scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
data_set1=sc.fit_transform(data_set)
```

```python
data_set1
```

```python
# elbow graaph for wcss value
from sklearn.cluster import KMeans
```

```python
wcss=[]

for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit_predict(data_set1)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.axvline(x=4,linestyle="--",color="green")
plt.title("elbow graph")
plt.xlabel("no of cluster")
plt.ylabel("wcss")
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(4)
pred1=km.fit_predict(data_set1)
pred1
```

```python
# centre of cluster 
```

```python
km.cluster_centers_
```

```python
data_set["cluster"]=pred1
```

```python
data_set
```

```python
data_set.groupby("cluster").agg("mean")
```

# DBSCAN

```python
Data=pd.read_csv("crime_data.csv")
```

```python
Data
```

```python
Data.drop(columns="Unnamed: 0",inplace=True)
```

```python
Data
```

```python
Data.describe()
```

```python
Data.boxplot()
```

```python
# featurer scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
Data1=sc.fit_transform(Data)
Data1
```

```python
# finding epsilon value
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neighbor=NearestNeighbors(n_neighbors=3).fit(Data1)
neighbor
```

```python
d,index=neighbor.kneighbors(Data1)
d
```

```python
distance=np.sort(d[:,1])
distance
```

```python
plt.plot(distance)
plt.xlabel("index")
plt.ylabel("eps value")
plt.title("k-distance graph")
```

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN(eps=1.30,min_samples=4)
predictions=db.fit_predict(Data1)
```

```python
predictions
```

```python
plt.scatter(Data1[:,0],Data1[:,3],c=predictions,cmap="viridis")
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(Data1,predictions)
```

```python
# analyze cluster
```

```python
Data["cluster"]=predictions
```

```python
Data
```

```python
Data.groupby('cluster').agg("mean")
```
