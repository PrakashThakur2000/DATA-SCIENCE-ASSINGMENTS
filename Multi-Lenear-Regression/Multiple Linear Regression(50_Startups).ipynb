# TRANFORMATION

```python
TR=pd.get_dummies(data=df,columns=['State'])
TR
```

```python
df.rename(columns={'R&D Spend':'RDSpend'},inplace=True)
```

```python
df.rename(columns={'Marketing Spend':'MSpend'},inplace=True)
```

```python
from sklearn .preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
sc.fit_transform(TR)
```

```python
df2=pd.DataFrame(sc.fit_transform(TR),columns=TR.columns)
df2
```

# Model building

```python
import statsmodels.formula.api as smf
```

```python
df3=df.drop(columns=['State'])
```

```python
df3
```

```python
Scaled_data=pd.DataFrame(sc.fit_transform(df3),columns=df3.columns)
Scaled_data
```

```python
model=smf.ols('Profit~RDSpend+Administration+MSpend',data=Scaled_data).fit()
model
```

```python
model.params
```

```python
model.rsquared
```

```python
model.rsquared_adj
```
