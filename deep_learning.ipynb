```python
#import libraries
!pip install tensorflow
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#load the data
```

```python
df=pd.read_csv('/Churn_Modelling.csv')
df
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
df
```

```python
df=pd.get_dummies(data=df,columns=['Geography','Gender'])
df
```

model bulding

```python
sc=StandardScaler()
x=df.drop(columns=['Exited'])
y=df['Exited']
```

```python
x=sc.fit_transform(x)
x
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#initialise model
ann=Sequential()
#add hiden layer
ann.add(Dense(units=10,activation='relu'))
#add output layer
ann.add(Dense(units=1,activation='sigmoid'))
#establish the connection between the layer
ann.compile(optimizer='adadelta',loss='binary_crossentropy',metrics='accuracy')
#fit the data
ann.fit(xtrain,ytrain,epochs=100)
```

```python
ypred=ann.predict(xtest)
ypred
```

```python
ypred>0.5
```

```python
ypred=ypred>0.5
ypred
```

```python
ytest
```

```python
print(classification_report(ytest,ypred))
```

Tunning the ann model

```python
import keras_tuner as kt
```

```python
def optimizer_selection(hp):
  model = Sequential()
#add hiden layer
  model.add(Dense(units=10,activation='relu'))
#add outeput layer
  model.add(Dense(units=1,activation='sigmoid'))
#optimizer selection
  optime=hp.Choice('optimizer',values=['sgd','adam','rmsprop'])
  model.compile(optimizer=optime,loss='binary_crossentropy',metrics='accuracy')
  return model
```

```python
tuner=kt.RandomSearch(
    optimizer_selection,
    objective='val_accuracy',
    max_trials=3
)
```

```python
tuner.search(xtrain,ytrain, epochs=3,validation_data=(xtest,ytest))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model=tuner.get_best_models(num_models=1)[0]
model.fit(xtrain,ytrain,epochs=100,validation_data=(xtest,ytest))
```
